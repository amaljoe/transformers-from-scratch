```python
from transformers import AutoModel, AutoProcessor, Qwen3VLForConditionalGeneration
from chandra.model.hf import generate_hf
from chandra.model.schema import BatchInputItem
from chandra.output import parse_markdown
import os
import torch

model_name = '/home/compiling-ganesh/24m0797/.cache/huggingface/hub/models--datalab-to--chandra/snapshots/b2c94232ad5b54e82eda20a925a7cad165a8d603'

os.environ['CUDA_VISIBLE_DEVICES'] = '0,1,2,3'

model = Qwen3VLForConditionalGeneration.from_pretrained(model_name, 
                                                        device_map="auto",
                                                        torch_dtype=torch.bfloat16)
model.processor = AutoProcessor.from_pretrained(model_name)
```

```python
from chandra_input import load_pdf_images

images = load_pdf_images('data/gst_annual.pdf', [43])

batch = [
    BatchInputItem(
        image=PIL_IMAGE,
        prompt_type="ocr_layout"
    ) for PIL_IMAGE in images
]

result = generate_hf(batch, model)[0]
markdown = parse_markdown(result.raw)
```
